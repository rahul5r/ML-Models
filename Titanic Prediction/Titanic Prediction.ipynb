#### **Loading the Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
# SibSp : No. of siblings/spouses
# Parch : No. of parents / children
X = df[["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare"]]
y = df["Survived"]
```

```python
X['Sex'] = X['Sex'].replace({'male': 1, 'female': 0, 'other': 2})
```

```python
X.head()
```

```python
null_count = X.isnull().sum()
null_count
```

```python
X["Age"] = X["Age"].interpolate(method="linear")
null_count = X.isnull().sum()
null_count
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(X_train.loc[0])
print(y_train.loc[0])
```

#### **Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
```

```python
linear_model.fit(X_train, y_train)
```

```python
linear_model.score(X_test, y_test)
```

#### **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(max_iter=1000, random_state=42)
```

```python
logistic_model.fit(X_train, y_train)
```

```python
logistic_model.score(X_test, y_test)
```

**From the above observation we see that Logistic Regression performs better in this case**

#### **Saving the model**

```python
import pickle

with open('TitanicPrediction_Model.pkl', 'wb') as f:
    pickle.dump(logistic_model, f)
```

#### **Loading the Model**

```python
from joblib import load

model = load('TitanicPrediction_Model.pkl')
```
